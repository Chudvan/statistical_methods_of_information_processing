### Оценивание параметров нелинейного объекта методом наименьших квадратов

Уравнение исследуемого в лабораторной работе объекта имеет вид:
    \begin{equation}y=f(t,A,B,C,D,E)=A\exp(Bt)\cos(Ct+D)+E+\eta\end{equation}
Уравнение модели объекта:
    \begin{equation}\tilde{y}=\tilde{A}\exp(\tilde{B}t)\cos(\tilde{C}t+\tilde{D})+\tilde{E}\end{equation}
Видно, что объект нелинеен по параметрам $A,B,C,D,E$. 
Данные параметры подлежат оцениванию, $t$ — единственная предикторная переменная, а $\eta$ — случайная ошибка, удовлетворяющая
обычным предположениям: $ E(\eta) = 0, D(\eta)= \sigma^2 $.

Найти оценки параметров $A,B,C,D,E$ предлагается с помощью применения МНК к линеаризованному объекту.

Для линеаризации объекта нужны частные производные от уравнения модели объекта по каждому из параметров.

Исходя из графика и уравнения объекта, подберите начальное приближение:
$A_{0},B_{0},C_{0},D_{0},E_{0}$

ПРИМЕЧАНИЕ: если в процессе оценивания параметров МНК разойдется, то возможно стоит выбрать более точное начальное приближение.

```python
#Ваш код здесь
A_0 = 1.5
B_0 = -math.log(2)/13
C_0 = 2*3.14/14
D_0 = 3.14/10 
E_0 =-0.2
print(B_0)
def Y(t, A_0, B_0, C_0, D_0, E_0):
    return (A_0*np.exp(B_0*t)*np.cos(C_0*t+D_0)+E_0)
```

Определите функции, возвращающие значения частных производных:

```python
#Ваш код здесь
#def dFdA(...):
#    return ...
def dFdA(t, A_0, B_0, C_0, D_0, E_0):
    y_A=np.exp(B_0*t)*np.cos(C_0*t+D_0)
    return y_A

def dFdB(t, A_0, B_0, C_0, D_0, E_0):
    y_B=A_0*t*np.exp(B_0*t)*np.cos(C_0*t+D_0)
    return y_B

def dFdC(t, A_0, B_0, C_0, D_0, E_0):
    y_C=-A_0*t*np.exp(B_0*t)*np.sin(C_0*t+D_0)
    return y_C

def dFdD(t, A_0, B_0, C_0, D_0, E_0):
    y_D=-A_0*np.exp(B_0*t)*np.sin(C_0*t+D_0)
    return y_D

def dFdE(t, A_0, B_0, C_0, D_0, E_0):
    y_E=1
    return y_E 
```

Реализуйте МНК для линеаризованного объекта и найдите оценку параметров: $\tilde{A},\tilde{B},\tilde{C},\tilde{D},\tilde{E}$

```python
#Ваш код здесь
c_j_new = np.array([A_0, B_0, C_0, D_0, E_0]) 
c_j_old = np.array([0]*5) 
c_j_new.shape = (5, 1) 
c_j_old.shape = (5, 1)
e = np.array(range(100), dtype = np.float) #массив значений "выход" - "выход модели"
e.shape = (100, 1)
j = 0

while (np.abs(c_j_new[0]-c_j_old[0])>0.00001) and (np.abs(c_j_new[1]-c_j_old[1])>0.00001) and (np.abs(c_j_new[2]-c_j_old[2])>0.00001) and (np.abs(c_j_new[3]-c_j_old[3])>0.00001) and (np.abs(c_j_new[4]-c_j_old[4])>0.00001):
    DF = np.zeros((100, 5)) 
    c_j_old = c_j_new
    for i in range(100):
        e[i] = data.U[i] - Y(data.t[i], c_j_old[0], c_j_old[1], c_j_old[2], c_j_old[3], c_j_old[4])
        DF[i][0]= dFdA(data.t[i], c_j_old[0], c_j_old[1], c_j_old[2], c_j_old[3], c_j_old[4])
        DF[i][1]= dFdB(data.t[i], c_j_old[0], c_j_old[1], c_j_old[2], c_j_old[3], c_j_old[4])
        DF[i][2]= dFdC(data.t[i], c_j_old[0], c_j_old[1], c_j_old[2], c_j_old[3], c_j_old[4])
        DF[i][3]= dFdD(data.t[i], c_j_old[0], c_j_old[1], c_j_old[2], c_j_old[3], c_j_old[4])
        DF[i][4]= dFdE(data.t[i], c_j_old[0], c_j_old[1], c_j_old[2], c_j_old[3], c_j_old[4])
        
    R=np.eye(100)
    DF = np.asmatrix(DF)
    e = np.asmatrix(e)
    c_j_new = c_j_old + (DF.getT() * R * DF).getI() * DF.getT() * R * e #уточнение оценки на каждой итерации
    print("Итерация №", j)
    print("A = ", c_j_new[0])
    print("B = ", c_j_new[1])
    print("C = ", c_j_new[2])
    print("D = ", c_j_new[3])
    print("E = ", c_j_new[4])
    j = j + 1
```

Постройте графики для исходных измерений и полученной модели объекта:

```python
print("Количество итераций = ", j) 
print("Исходный параметр A = ", A_0, "  Параметр модели = ", c_j_new[0])
print("Исходный параметр B = ", B_0, "  Параметр модели  = ", c_j_new[1])
print("Исходный параметр C = ", C_0, "  Параметр модели  = ", c_j_new[2])
print("Исходный параметр D = ", D_0, "  Параметр модели  = ", c_j_new[3])
print("Исходный параметр E = ", E_0, "  Параметр модели  = ", c_j_new[4])
```

```python
plt.figure(figsize=(15, 10))
plt.plot(data.t, data.U, color='red')
plt.grid(True)

#График для полученной модели
y = numpy.array(range(100), dtype = np.double)
y.shape = (100,1) 
i=0
for i in range(100):
    y[i]=c_j_new[0]*np.exp(c_j_new[1]*i)*np.cos(c_j_new[2]*i+c_j_new[3])+c_j_new[4]
    i=i+1
plt.plot(data.t, y, color='blue')
plt.grid(True)
```

Постройте график остатков и сделайте выводы о применимости линеаризации для оценки параметров нелинейных объектов:
